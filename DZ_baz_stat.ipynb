Из первичного анализа столбца "rectal temperature" можно сделать выводы:
- данные корректные, выбросов нет
- стандартное отклонение не большое
- согласно описания данные зна

```python
q1 = df['pulse'].quantile(0.25)
q3 = df['pulse'].quantile(0.75)
iqr = q3 - q1
lower_bound = q1 - (1.5 * iqr) 
upper_bound = q3 + (1.5 * iqr)
remove_outliers = df[~df['pulse'].between(lower_bound, upper_bound, inclusive=True)]
remove_outliers
```

Из значений выбросов по показателю Пульс, мы видим что они присущи молодым лошадям, что показывет есть статистически значимая зависимость между пульсом и возростом животного

```python
df['pain'] = df['pain'].fillna(df['pain'].round().mode()[0])
df.head(20)
```

Дополняем данные в столбец 'temperature of extremities', исходя из описания, данный показатель зависит от 'rectal temperature', при повышенной температуре, повышантся температура конечностей (значение 4)

```python
df.loc[df['rectal temperature']>38, ['temperature of extremities']] = df['temperature of extremities'].fillna(4)
df.isnull().sum()
```

Принятие решения о заполнении недостающих данных необходимо принимать исходя из поставленных задач оналитики.
- Допустим значение столбца 'pain' можно заполнить произвольными данными, если нам надо узнать, допустим эффективность проведения операций.
- значение столбца temperature of extremities можно заполнить исходя из данных rectal temperature, при повышенной температуре лошади - будет значение 1 
- значение puls можно заролнить для лошадей испытывающих шок или болезненные ощущения (значение столбца pain)
- значение respiratory rate имеет сомнительную значимость и может быть заполнени средними значениями
- значение rectal temperature имеет очень важные паказатели для выявления состояния лошади и его заполнение может сильно исказить данные, но для анализа появения болезни у лошадей в зависиммости от возраста не будет имет значения и тогда его можно не учитывать
