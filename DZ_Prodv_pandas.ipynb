```python
import pandas as pd
```

```python
log = pd.read_csv('/Users/admin/Desktop/Учеба/Docs/visit_log.csv', sep=';')
log.head()
```

```python
def filtr (data):
    if data['traffic_source'] == 'yandex' or data['traffic_source'] == 'google':
        var = 'organic'
    elif (data['traffic_source'] == 'paid' or data['traffic_source'] == 'email') and data['region'] == 'Russia':
        var = 'ad'
    elif (data['traffic_source'] == 'paid' or data['traffic_source'] == 'email') and data['region'] != 'Russia':
        var = 'other'
    else:
        var = data['traffic_source']
    return var
```

```python
log['source_type'] = log.apply(filtr, axis=1)
log.head(10)
```

### Задание 3

```python
import pandas as pd
```

```python
df = pd.read_csv('/Users/admin/Desktop/Учеба/Docs/ml-latest-small/ratings.csv')
df.head()
```

```python
count = df['userId'].value_counts().reset_index()
count.columns = ['userId', 'count']
count.head()
```

```python
df = df.merge(count, on ='userId')
```

```python
df = df[df['count'] > 100]
df.head()
```

```python
def time_(row):
    return row.timestamp.max() - row.timestamp.min()
```

```python
df.groupby('userId').apply(time_).mean()
```
