```python
documents = [
{'type': 'passport', 'number': '2207 876234', 'name': 'Василий Гупкин'},
{'type': 'invoice', 'number': '11-2', 'name': 'Геннадий Покемонов'},
{'type': 'insurance', 'number': '10006', 'name': 'Аристарх Павлов'}
]
directories = {
'1': ['2207 876234', '11-2'],
'2': ['10006'],
'3': []
}

def p(doc):
    '''
    Функция выводит Имя по Номеру документа
    '''
    for number in documents:
        if doc == number['number']:
            return f'Владелец документа:{number["name"]}'
    else: 
        return 'Документ не найден в базе'

def s(doc):
    '''
    Функция выводит Номер полки по Номеру документа
    '''
    for k_num, number in directories.items():
        for num in number:
            if doc == num:
                return f'Документ хранится на полке:{k_num}'
        return 'Документ не найден в базе'

def l():
    '''
    Функция выводит инвормацию по все документам и на какой полке находится
    '''
    for k_num, number in directories.items():
        for sp in documents:            
            if sp['number'] in number:
                print(f"№:{sp['number']}, тип:{sp['type']}, владелец:{sp['name']}, полка хранения:{k_num}")
                    
def ads(n_polki):   
    if n_polki not in directories:
        directories[n_polki] = []
        return f'Полка добавлена. Текущий перечень полок: {", ".join(map(str, directories))}'
    else: 
        return f'Такая полка уже существует. Текущий перечень полок: {", ".join(map(str, directories))}'

def ds(n_polki):
    if n_polki in directories:    
        if len(directories[n_polki]) == 0:
            del directories[n_polki]
            return f'Полка удалена. Текущий перечень полок: {", ".join(map(str, directories))}'
        else: 
            return f'На полке есть документа, удалите их перед удалением полки. Текущий перечень полок: {", ".join(map(str, directories))}'
    else:
        return f'Такой полки не существует. Текущий перечень полок: {", ".join(map(str, directories))}'

def ad(doc, tip, vlad, n_polki):
    if n_polki not in directories:
        return print(f'Такой полки не существует. Добавьте полку командой as. \nТекущий список документов:')
    else:
        documents.append({'type':tip, 'number':doc, 'name':vlad})
        directories[n_polki] = doc
        return print(f'Документ добавлен. Текущий список документов:')

def d(doc, documents):
    new_document = []
    counter = 0
    for str in documents:
        if doc == str['number']:
            del documents[counter]
            print(f'Документ удален. \nТекущий список документов:')
            l()
            return documents
        else: 
            counter += 1
    return print(f'Такой документ не найден. \nТекущий список документов:'), l()

def main():
    while True:
        comand = input('Введите команду')
        if comand == 'p':
            doc = input('Введите номер документа')
            print(p(doc))
        elif comand == 's':
            doc = input('Введите номер документа')
            print(s(doc))
        elif comand == 'l':
            l()
        elif comand == 'ads':
            n_polki = input('Введите номер полки:')
            print(ads(n_polki))
        elif comand == 'ds':
            n_polki = input('Введите номер полки:')
            print(ds(n_polki))
        elif comand == 'ad':
            doc = input('Введите номер документа')
            tip = input('Введите тип документа:')
            vlad = input('Введите владельца документа:')
            n_polki = input('Введите полку для хранения:')
            ad(doc, tip, vlad, n_polki)
            l()
        elif comand == 'd':
            doc = input('Введите номер документа')
            (d(doc, documents))  
        elif comand == 'q': 
            return

main()



```
