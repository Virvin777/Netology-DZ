```python
import pandas as pd
```

```python
movies = pd.read_csv('movies.csv')
```

```python
ratings = pd.read_csv('ratings.csv')
```

```python
data = movies.merge(ratings, how = 'left') # Объеденяем таблицы с названием фильмов и оценками
```

```python
data = data[['title', 'rating']]
```

```python
def class_(data):
    if data['rating'] <= 2:
        return 'низкий рейтинг'
    elif data['rating'] <= 4:
        return 'средний рейтинг'
    else: return 'высокий рейтинг'
```

```python
data = data.groupby('title').mean()
```

```python
data['class'] = data.apply(class_, axis = 1)
```

```python
df = pd.read_csv('keywords.csv')
```

```python
df
```

```python
geo_data = {
'Центр': ['москва', 'тула', 'ярославль'],
'Северо-Запад': ['петербург', 'псков', 'мурманск'],
'Дальний Восток': ['владивосток', 'сахалин', 'хабаровск']
}
```

```python
def region(data):
    zapros = data.split(' ')
    for word in zapros:
        if word in geo_data.values():
            for reg, city in geo_data.items():
                if city == word:
                    return reg
    return 'undefined'
```

```python
df['region'] = df.apply(region, axis =1)
```
